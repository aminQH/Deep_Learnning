```python
import tensorflow as tf
from tensorflow import keras
```

```python
# y = x * 3 - 60 / 2
x = tf.constant([12 , 15 , 16 , 17 , 19 , 20] , dtype=float)
y = x * 3 -60 /2
```

```python
model = keras.Sequential([keras.layers.Dense(units= 1 , activation=None ,input_shape =[1] )])
model.compile(optimizer='adam' ,loss='mse')
model.summary()
```

```python
model.fit(x,y , batch_size=1 , epochs=600)
```

```python
x_test = tf.constant([99 , 98 , 100 , 200 , 150 , 11 , 17] , dtype=float)
y_test = x_test * 3 -60 /2

y_predict = model.predict(x_test)
```

y = x^2 + 5 * x + 3

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
```

```python
x = np.arange(-80 , 81 , 0.5  )
y = x**2 + 5 * x + 3
```

```python
x_trin , x_test , y_train , y_test = train_test_split(x ,y , test_size=0.2)
```

# Model_defenition

```python
model_1 = keras.Sequential()
model_1.add(keras.layers.Dense(units=1024 , activation='relu' , input_shape=[1]))
model_1.add(keras.layers.Dense(units = 1))
```

```python
model_1.compile(optimizer='adam' , loss='mse')
model_1.summary()
```

```python
model_1.fit(x_trin , y_train , batch_size=1 , epochs=1000 , validation_data=(x_test , y_test))
```
